```python
# Import required libraries
import pandas as pd
import numpy as np

# Create a sample DataFrame with reviews
sample_reviews = [
    "This product is amazing! I love it so much!!!",
    "Not very good quality, wouldn't recommend. :(",
    "Pretty decent product for the price. Worth buying.",
    "The customer service was excellent! 5 stars!",
    "Average product, nothing special about it..."
]

# Create DataFrame
df = pd.DataFrame({
    'review': sample_reviews
})

# Display the DataFrame
print("Sample DataFrame created with reviews:")
display(df)
```

```python
# Import required NLTK libraries
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import re

# Download required NLTK data
print("Downloading required NLTK data...")
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

# Initialize lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    """
    Preprocess text by performing the following steps:
    1. Convert to lowercase
    2. Remove special characters and numbers
    3. Tokenize
    4. Remove stopwords
    5. Lemmatize
    """
    try:
        # Convert to string if not already
        text = str(text)
        
        # Convert to lowercase
        text = text.lower()
        
        # Remove special characters and numbers
        text = re.sub(r'[^a-zA-Z\s]', '', text)
        
        # Tokenize
        tokens = word_tokenize(text)
        
        # Remove stopwords and lemmatize
        cleaned_tokens = [lemmatizer.lemmatize(token) for token in tokens if token not in stop_words]
        
        # Join tokens back into text
        cleaned_text = ' '.join(cleaned_tokens)
        
        return cleaned_text
    
    except Exception as e:
        print(f"Error processing text: {str(e)}")
        return text  # Return original text if processing fails

print("Text preprocessing function is ready!")
```

```python
# Apply preprocessing to the reviews
print("Preprocessing the reviews...")
df['cleaned_text'] = df['review'].apply(preprocess_text)
print("Preprocessing complete!")

# Display a comparison of original vs cleaned reviews
print("\nComparison of original vs cleaned reviews:")
display(df[['review', 'cleaned_text']])

# Display some statistics
print("\nText cleaning statistics:")
print(f"Average original review length: {df['review'].str.len().mean():.1f} characters")
print(f"Average cleaned review length: {df['cleaned_text'].str.len().mean():.1f} characters")
```
